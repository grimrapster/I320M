```python
# Import all the libraries to test installation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset
df = pd.read_csv('BreastCancerWisconsin/data.csv')
```

```python
# Display basic information
print("=" * 50)
print("Dataset Shape:", df.shape)
print("=" * 50)
print("\nFirst few rows:")
print(df.head())

print("\n" + "=" * 50)
print("Dataset Info:")
print("=" * 50)
df.info()

print("\n" + "=" * 50)
print("Diagnosis Distribution:")
print("=" * 50)
print(df['diagnosis'].value_counts())
```

```python
# Basic statistics
print("\n" + "=" * 50)
print("Summary Statistics:")
print("=" * 50)
print(df.describe())
```

```python
# Test matplotlib - simple bar plot
plt.figure(figsize=(6, 4))
df['diagnosis'].value_counts().plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis (B=Benign, M=Malignant)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('diagnosis_distribution.png', dpi=100)
print("\n✓ Matplotlib test passed - saved 'diagnosis_distribution.png'")
```

```python
# Test seaborn - correlation heatmap
plt.figure(figsize=(8, 6))
numeric_cols = df.select_dtypes(include=[np.number]).columns[:5]
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap (First 5 Features)')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=100)
print("✓ Seaborn test passed - saved 'correlation_heatmap.png'")
```

```python
# Test scipy - simple statistical test
if 'radius_mean' in df.columns:
    benign = df[df['diagnosis'] == 'B']['radius_mean']
    malignant = df[df['diagnosis'] == 'M']['radius_mean']
    t_stat, p_value = stats.ttest_ind(benign, malignant)
    print(f"\n✓ Scipy test passed - T-test result: t={t_stat:.3f}, p={p_value:.6f}")
```
